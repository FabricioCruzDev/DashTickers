```python
import pandas as pd
import yfinance as yf
```

```python
pd.set_option('display.max_columns', 50)
```

```python
my_tickers = ['ITUB3.SA', 'AMER3.SA', 'BBAS3.SA', 'CMIG4.SA', 'CXSE3.SA', 'GOAU4.SA', 'ITSA4.SA', 'RAIZ4.SA', 'TRPL4.SA', 'VALE3.SA', 'KEPL3.SA']
df = pd.DataFrame()

def download_data(ticker):
  df_teste = yf.Ticker(ticker)
  df_teste = df_teste.history(period='max')
  df_teste['Ticker'] = ticker
  return df_teste

for ticker in my_tickers:
  df_ticker = download_data(ticker).reset_index()
  df = pd.concat([df, df_ticker])
```

```python
df = df[['Date', 'Ticker', 'Open', 'High', 'Low', 'Close', 'Dividends']]
df['Date'] = pd.to_datetime(df['Date']).dt.date
df.head()
```

```python
df = df.apply(lambda x: x.apply(lambda y: '{:.4f}'.format(y) if isinstance(y, float) else y))
```

```python

df['Close'] = pd.to_numeric(df['Close'], errors='coerce')
df['Open'] = pd.to_numeric(df['Open'], errors='coerce')


df['Var'] = df['Close'] - df['Open']
df['%Var'] = (df['Var'] / df['Open'])
df.head()
```

```python
df.to_csv('df_tickers.csv')

##Em seguida atualizar o drive virtual e fazer o download do arquivo
```
